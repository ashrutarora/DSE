#### Filling the NULL values

```python
df1 = df.fillna(value = '0')
df1.head()
```

#### Alternative method --> Puts the previous rows value for the missing value

```python
df2= df.fillna(method = 'pad')
df2.head()
```

#### Alternative method --> Puts the next rows value for the missing value

```python
df3 = df.fillna(method = 'bfill')
df3.head()
```

#### Fills the NULL values with the PREVIOUS Column values

```python
df4 = df.fillna(method = 'pad', axis = 1)
df4.head()
```

#### Fills the NULL values with the NEXT Column values

```python
df5 = df.fillna(method = 'bfill', axis = 1)
df5.head()
```

#### Filling the missing values in separate columns with the values we specify

```python
df6 = df.fillna({
    'society': 'abcd',
    'balcony': 'defg'
})
df6.head()
```

#### Filling the NULL values with MEAN of that column

```python
df7 = df.fillna(value = df['balcony'].mean())
df7.head()
```

#### Filling the NULL values with MAX of that column

```python
df8 = df.fillna(value = df['balcony'].max())
df8.head()
```

#### Drop the rows that contain NULL values

```python
df9 = df.dropna()
df9.head()
```

#### Creating a table with NOT NULL values from Balcony column

```python
df10 = df.loc[df['balcony'].isnull()== False]
df10.head()
```

#### Creating a table with NULL values from Balcony column

```python
df11 = df.loc[df['balcony'].isnull()== True]
df11.head()
```

```python
# Lurking Variables
```

```python
# Dropping rows that contain NULL values using "HOW"
# How takes 2 values --> ANY , ALL
# Any : If any one of the row values contain NULL, then that row is dropped
# All : If all the row values contain NULL, then only those rows are dropped

df12 = df.dropna(how = 'any')
df12
```

```python
df13 = df.dropna(how = 'all')
df13
```

#### Replacing the NULL values with a DIFFERENT value

```python
df14 = df.replace(to_replace = np.nan, value = 875465)
df14

# Replaces all the null values in the dataset with 875465
```

#### Replacing specific values with values specified

```python
df15 = df.replace(to_replace = 3.0, value = 5.0)
df15

# Wherever there is 3.0 value, it is replaced by 5.0
```
