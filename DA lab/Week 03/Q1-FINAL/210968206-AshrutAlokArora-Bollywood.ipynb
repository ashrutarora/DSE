```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("bollywood.csv")
```

```python
df
```

```python
# 1. How many records are present in the dataset?
df.count()
```

```python
df2 = df.groupby(['Genre'])['Genre'].count()
print(df2)
```

```python
df.Genre.value_counts()
```

```python
# 2. How many movies got released in each genre? Sort number of releases in each genre
# in descending order.

df["Genre"] = df.Genre.str.strip()
df.Genre.value_counts()
```

```python
# 3. Which genre had highest number of releases?

print("Highest number of releases: ")
df2.idxmax()
```

```python
#  4. How many movies in each genre got released in different release times like long
#    weekend, festive season, etc. (Note: Do a cross tabulation between Genre
#    and ReleaseTime.)

print("Crosstab between Genre and ReleaseTime")

pd.crosstab(df.Genre, df.ReleaseTime)
```

```python
df['Release Date'] = pd.to_datetime(df['Release Date'])
df
```

```python
df['Month'] = df['Release Date'].dt.month
df
```

```python
df3 = df.groupby(['Month'])['Month'].count()
df3
```

```python
#  5. Which month of the year, maximum number movie releases are seen?

print("Month with MAX number of movies: ") 
df3.idxmax()
```

```python
# 6. Which month of the year typically sees most releases of high budgeted movies, that is,
#    movies with budget of 25 crore or more?


print("Month with movies that have Budget 25cr or more: ")

df.query('Budget>=25').Month.value_counts().idxmax()
```

```python
df3 = df['BoxOfficeCollection'].subtract(df["Budget"])
df['ROI'] = df3.divide(df['Budget'])
df
```

```python
# 7. Which are the top 10 movies with maximum return on investment (ROI)?

df.nlargest(10, ['ROI'])
```

```python
df.ReleaseTime.value_counts()
```

```python
#  8. Do the movies have higher ROI if they get released on festive seasons or long
#     weekend? Calculate the average ROI for different release times.

df.groupby('ReleaseTime', as_index=False)['ROI'].mean()
```

```python
# 9. Is there a correlation between box office collection and YouTube likes? Is the
#    correlation positive or negative?

boxplot = df.boxplot(column=['BoxOfficeCollection', 'YoutubeLikes'])
```

```python
# 10. Which genre of movies typically sees more YouTube likes? Draw boxplots for each
#     genre of movies to compare.

sns.boxplot(data=df,y='Genre',x='YoutubeLikes')
```

```python
# 11. Which of the variables among Budget, BoxOfficeCollection, YoutubeView,
#     YoutubeLikes, YoutubeDislikes are highly correlated? Note: Draw pair plot
#     or heatmap.


sns.pairplot(df)
```

```python
# HEATMAP

df_corr = ['Budget','BoxOfficeCollection','YoutubeViews','YoutubeLikes','YoutubeDislikes']
sns.heatmap(df[df_corr].corr(),annot=True);
```

```python
# 12. During 2013−2015 period, highlight the genre of movies and their box office
#     collection? Visualize with best fit graph.


df5 = df.loc[(df["Release Date"] >= "2013-01-01") & (df["Release Date"] <= "2015-12-31")]

sns.barplot(x = 'Genre',
            y = 'BoxOfficeCollection',
            data = df5)
```

```python
df['Year'] = df['Release Date'].dt.year

df6 = df.groupby(['Year'])['Year'].count()
df6
```

```python
sns.barplot(x='Year', y='BoxOfficeCollection', hue='Genre', data=df, ci=None)
```

```python
# 13. Visualize the Budget and Box office collection based on Genre

plt.bar(df['Genre'],df['BoxOfficeCollection'],align='edge',label='BoxOfficeCollection',width=0.5)
plt.bar(df['Genre'],df['Budget'],label='Budget')
```
