```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

### Generating Random Datasets

```python
import random

dataset1 = []
dataset2 = []

for i in range(25):
    dataset1.append(random.randint(1, 100))

for i in range(25):
    dataset2.append(random.randint(1, 100))

print(dataset1)
```

### b) Bar charts

### Normal Bar Chart

```python
plt.bar(dataset1, dataset2, width = 1.5, edgecolor = 'black', color = "green")

plt.title('Normal Bar Chart')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.show()
```

### Side by Side Bar Chart

```python
x1 = np.arange(len(dataset1))
w = 0.3

plt.bar(x1[:5], dataset1[:5], width = w, edgecolor = 'black')
x1 = x1+w
plt.bar(x1[:5], dataset2[:5], width = w, edgecolor = 'black')

plt.title('Side by Side Bar Chart')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
```

### Stacked Bar Chart

```python
x1 = np.arange(len(dataset1))
plt.bar(x1, dataset1, label = '1', edgecolor = 'black')
plt.bar(x1, dataset2, bottom = dataset1, label = '2', edgecolor = 'black')

plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Stacked Bar Chart')

plt.legend()
plt.show()
```

### Another Example

```python
genres = ['Action', 'Thriller', 'Romance', 'Drama', 'SciFi']

boys = []
for i in range(5):
    boys.append(random.randint(1, 75))
    
girls = []
for i in range(5):
    girls.append(random.randint(1, 75))
```

```python
plt.bar(genres, boys, label = 'Boys', edgecolor = 'black')
plt.bar(genres, girls, bottom = boys, label = 'Girls', edgecolor = 'black')

plt.xlabel('Genres')
plt.ylabel('No of people')
plt.title('Fav types of Movies')

plt.legend()
plt.show()
```

```python
x1 = np.arange(len(genres))
w = 0.3

plt.bar(x1, boys, w, edgecolor = 'black')
x1 = x1+w
plt.bar(x1, girls, w, edgecolor = 'black')

plt.xlabel('Genres')
plt.ylabel('No of people')
plt.title('Fav types of Movies')

plt.xticks(x1-0.15, genres)
plt.show()
```

### d) 3D Scatter and Surface plots using matplotlib and seaborn packages.

```python
fig = plt.figure(figsize=(5,5))
ax = fig.add_subplot(projection='3d')

sns.set_style("ticks")
ax.scatter(dataset1,dataset2)

ax.set_xlabel("X AXIS")
ax.set_ylabel("Y AXIS")
ax.set_zlabel("Z AXIS")
ax.set_title('3D Scatter Plot')

plt.show()
```

```python
d1 = []
d2 = []
d1.append(np.arange(-1,1,0.02))
d2.append(np.arange(-1,1,0.02))

d1, d2 = np.meshgrid(d1, d2)
d3 = np.sin(np.sqrt(d1**2 + d2**2))

fig = plt.figure(figsize=(5,5))
ax = fig.add_subplot(projection='3d')
surf = ax.plot_surface(d1, d2, d3, cmap='plasma')

ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_zlabel('Z Axis')
ax.set_title('3D Surface Plot')

fig.colorbar(surf)

plt.show()
```

### Q3. Draw the Box Plots for five number summary for numerical columns of iris and mtcars dataset.

```python
from sklearn.datasets import load_iris

iris = load_iris()
iris = pd.DataFrame(data=iris.data, columns=iris.feature_names)

iris.head()
```

```python
fig, ax = plt.subplots(figsize=(8, 6))

sns.boxplot(data=iris, ax=ax, linewidth=2, fliersize=4)

ax.set_xlabel('Features', fontsize=14)
ax.set_ylabel('Values', fontsize=14)
ax.set_title('Boxplot of Iris Features', fontsize=16)

# Show the plot
plt.show()
```

### Q4. Create dataset of student marks in 5 subjects program wise (CSE, ECE, EEE, ICT programs) with for atleast 10 students each. Draw all suitable charts for this data.

```python
marks = {
    'CSE': np.random.randint(0, 101, size=10*5).reshape(10, 5),
    'ECE': np.random.randint(0, 101, size=10*5).reshape(10, 5),
    'EEE': np.random.randint(0, 101, size=10*5).reshape(10, 5),
    'ICT': np.random.randint(0, 101, size=10*5).reshape(10, 5),
}

marks
```

```python
average_marks = [np.mean(marks[program]) for program in marks]

plt.bar(marks.keys(), average_marks)

plt.title('Average Marks by Program')
plt.xlabel('Program')
plt.ylabel('Average Mark')

plt.show()
```

```python
data = [marks[program].flatten() for program in marks]

plt.boxplot(data, labels=marks.keys())

plt.title('Marks in All Subjects by Program')
plt.xlabel('Program')
plt.ylabel('Mark')

plt.show()
```

```python
fig,ax = plt.subplots(figsize=(10, 6))

for program, data in marks.items():
    subject_averages = np.mean(data, axis=0)

    ax.plot(np.arange(1, 6), subject_averages, label=program)
    

ax.set_title('Trend of Marks in 5 Subjects by Program')
ax.set_xlabel('Subject')
ax.set_ylabel('Average Mark')
ax.legend()

plt.show()
```
